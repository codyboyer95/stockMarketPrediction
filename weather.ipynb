```python
import matplotlib
from matplotlib import pylab
import pandas as pd
```

```python
log_df = pd.read_csv("./stockMarketPrediction/new_york_city.csv",
                     names=['date_time','maxtempC','mintempC','totalSnow_cm','sunHour',
                            'uvIndex','moon_illumination','moonrise','moonset','sunrise',
                            'sunset','DewPointC','FeelsLikeC','HeatIndexC','WindChillC',
                            'WindGustKmph','cloudcover','humidity','precipMM','pressure',
                            'tempC','visibility','winddirDegree','windspeedKmph','location'],
                     skiprows=1,
                     usecols=['date_time','maxtempC','mintempC','FeelsLikeC'],
                     na_values=['-'], 
                     encoding = "ISO-8859-1")
```

```python
ax = log_df.plot(figsize=(9,6))
ax.set_ylabel("Temp In C")
ax.set_xlabel("Day")
ax.set_title("Min, Max, and Feels Like Temp Per Day")
```

```python
log_df['date_time'] = pd.to_datetime(log_df['date_time'])
per = log_df.date_time.dt.to_period("M")
g = log_df.groupby(per)
ax2 = g.mean().plot(figsize=(9,6))
ax2.set_ylabel("Temp In C")
ax2.set_xlabel("Month")
ax2.set_title("Min, Max, and Feels Like Temp Averages Per Month")
```
